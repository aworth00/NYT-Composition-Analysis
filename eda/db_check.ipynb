```python
def update_flesch_kincaid():
    '''
    This function is meant to append all rows in data base with a new variable that represents the flesch_kincaid reading level
    of the article being looked at. VIA MONGO DB
    '''
    cursor = db.articles.find()
    for doc in cursor:
        article_id=doc['_id']
        flesch_score = textstat.flesch_reading_ease(doc['body'])
        articles.update_one({'_id':article_id},{'$set':{'flesch_kincaid_score':int(flesch_score)}})
```

```python
def score_update(df):
    '''
    writing test flesch-kincaid-ease score inside jupyter notebook
    '''
    nan_list = []
    for row, article in enumerate(df['body']):
        if len(df['body'][row])>0:
            x = article
            try:
                df.loc[row,'flesch_kincaid']=textstat.flesch_reading_ease(x)
            except (ValueError):

                df.loc[row,'flesch_kincaid']=np.nan()
                nan_list.append(row)
        else:
            nan_list.append(row)
    if len(nan_list)>0:
        return nan_list
#score_update(tester)        
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_17['body'], df_17['new_desk'], test_size=0.30, random_state=42)
```

```python
model = MultinomialNB()
vec = TfidfVectorizer()
train_mod = make_pipeline(vec,model)
train_mod.fit(X_train,y_train)
train_mod.predict_proba(X_test)
y_pred = train_mod.predict(X_test)

print(train_mod.score(y_pred,y_test))
```
